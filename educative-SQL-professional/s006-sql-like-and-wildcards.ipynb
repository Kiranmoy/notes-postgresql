University administrator request:

1. I want to fetch records of students whose names start with "`A`". How can I do that?

2. How can I check if a student’s name ends with "`son`" or has a specific number of characters?

3. I want to search for students whose names contain the word "`John`", but it can be anywhere in their name, not necessarily at the beginning or end.

4. How can I check if a student’s name contains the word "`son`"?

5. I need to find all students whose names do not contain the word "`a`". Can you help me with that?

# `LIKE` operator and wildcards

The `LIKE` operator is specifically designed for pattern-based searches in SQL. It allows us to match partial strings using special symbols called wildcards:

1. **`%` (Percentage Wildcard)** → Represents zero or more characters
2. **`_` (Underscore Wildcard)** → Represents a single character

Using `LIKE` with these wildcards, we can now correctly filter names based on the university administrator’s request.

# Solutions

**Fetch all students whose names start with "`A`"**.

```sql
SELECT * FROM Student WHERE FullName LIKE "A%";
```



---

**Find students whose names end with "`son`"**.

```sql
SELECT * FROM Student WHERE FullName LIKE "%son";
```



---

**Find departments with seven-letter department names starting with "`B`"**.

```sql
SELECT * FROM Department WHERE DepartmentName LIKE "B______";
```



---

**Find students whose names contain "`son`"**.

```sql
SELECT * FROM Student WHERE FullName LIKE "%son%";
```



---

**Find students whose names do not contain "`a`"**.

```sql
SELECT * FROM Student WHERE FullName NOT LIKE "%a%"
```
