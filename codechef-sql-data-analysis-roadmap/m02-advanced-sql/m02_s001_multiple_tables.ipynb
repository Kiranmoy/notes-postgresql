# Combining tables with SQL Joins

```postgresql

-- JOIN or INNER JOIN
-- Returns matching row present in both tables
SELECT *
FROM employee
INNER JOIN department
ON employee.employee_id = department.employee_id;

-- LEFT JOIN, or LEFT OUTER JOIN
-- All rows of the LEFT table are included, and Columns from the RIGHT table will be filled with NULL values if no match is found.
SELECT *
FROM customer
LEFT JOIN order
ON customer.cust_id = order.cust_id;

-- RIGHT JOIN, or RIGHT OUTER JOIN
-- All rows of the RIGHT table are included, and Columns from the LEFT table will be filled with NULL values if no match is found.
SELECT *
FROM customer
RIGHT JOIN order
ON customer.cust_id = order.cust_id;

-- FULL OUTER JOIN or FULL JOIN = INNER JOIN + LEFT JOIN + RIGHT JOIN
SELECT * FROM student s
FULL OUTER JOIN course c
ON s.Course_id = c.Course_id;

-- SELF JOIN
SELECT
    e1.employee_name AS Employee,
    e2.employee_name AS Manager
FROM
    employees AS e1  -- Alias for employee
INNER JOIN
    employees AS e2  -- Alias for manager
ON
    e1.manager_id = e2.employee_id;

-- CROSS JOIN
 SELECT *
 FROM Manufacture
 CROSS JOIN Model

 SELECT St_Name, Course_Name FROM student, course;

```
