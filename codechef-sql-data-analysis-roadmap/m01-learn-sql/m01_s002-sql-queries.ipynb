# SELECT, DISTINCT, WHERE, BETWEEN

```sql
SELECT * FROM flights;
SELECT DISTINCT origin from flights;

SELECT * FROM flights
WHERE origin = 'Mumbai';

SELECT * FROM flights
WHERE passenger_id BETWEEN 10001 AND 10007;

```

# AND, OR, LIKE

```sql
SELECT * FROM flights
WHERE origin = 'Mumbai'
AND passenger_id BETWEEN 10001 AND 10007;

SELECT * FROM flights
WHERE origin = 'Mumbai'
OR origin = 'New York';

SELECT * FROM flights
WHERE passenger_id LIKE '1000_';

```

# ORDER BY, LIMIT, NULL

```sql
SELECT * FROM flights
ORDER BY passenger_name;

SELECT * FROM flights
ORDER BY passenger_name ASC;

SELECT * FROM flights
ORDER BY passenger_name DESC;

SELECT * FROM flights
ORDER BY passenger_name
LIMIT 3;

SELECT * FROM flights
WHERE origin IS NULL;

SELECT * FROM flights
WHERE origin IS NOT NULL;
```
