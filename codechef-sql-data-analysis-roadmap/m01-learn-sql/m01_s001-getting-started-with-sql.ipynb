# ER Diagram

An ER diagram is like a blueprint (visual representation) for your database, that:
* Shows how data is organized in a database.
* Maps out relationships between different types of entities.
* Helps you design databases before you actually build them.

# Components of ERD

* Entities
* Attributes
* Relationships



# Entity

An **entity** is a "thing" or "object" in the real world that can be identified and is distinguishable from other objects. In our fitness tracking example, the entities we worked with were: **Users**, **Workouts**, **Activities**, and **Goals**.

**Entity Set**: An Entity Set is a collection of similar entities that share the same attributes.

**Characteristics of an Entity Set**:
* Contains entities of the same type
* Shares a common structure (attributes)
* Each entity in the set is unique
* Can participate in relationships with other entity sets

# Type of entities

**Strong Entity**: 
* A Strong Entity is an entity that can exist independently of other entities.
* It has attributes that can uniquely identify it (like a primary key).
* Strong entities are represented by a rectangular box.

**Weak Entity**
* A Weak Entity is an entity that cannot exist independently and must be associated with another strong entity.
* It typically does not have a unique identifier of its own and relies on a foreign key to identify it.
* Shown as a rectangle with double borders.

# Attributes

* Attributes are the **properties** or **characteristics** of an entity.
* They provide more details about the entity, like specific values that describe or identify it.
* In simple terms, attributes are the **columns** in a table in a database, while the entity is the **table** itself.

## Key Attribute

* Key attribute is basically something unique that helps us identify each record distinctly.
* Key attribute is represented by an **oval with underlying lines**.

**Points to Remember**:
* Must be unique (no duplicates allowed).
* Cannot be null.
* Should rarely or never change.
* Often used as references in other tables.

## Composite Attribute

* A composite attribute is when you combine multiple simple attributes to form a single meaningful attribute.
* Composite attribute is represented by an **oval**.

**Points to Remember**:
* Can be broken down into smaller parts.
* Each part has meaning on its own.
* Can be used together or separately.
* Good for organizing related data.

# Relationships

* A relationship is a connection or association between specific entities.
* Relationship type is represented by a **diamond** and connecting the entities with lines.

**Relationship set**: A relationship set is a collection of similar relationships between entities of the same type.


# Constraints Overview

**Constraints** provide details about the usage of a column and are applied after specifying the column's data type.
They enable the database to reject any inserted data that violates a particular constraint.

**Primary Key**
* A **Primary Key** is a column (or a set of columns) in a database table that uniquely identifies each row.
* It enforces **uniqueness** and **non-nullability**, meaning no two rows can have the same primary key value, and it cannot be NULL.

**Unique**
* Ensures that all values in a column are distinct.
* Unlike a primary key, a column with a `UNIQUE` constraint can have `NULL` values (unless explicitly marked as `NOT NULL`).
* Example: `id INT NOT NULL`

**Not Null**
* Prevents a column from having `NULL` values.
* Ensures that every row must have a valid (non-null) value for the specified column
* Example: `id INT NOT NULL`

**Default**
* Assigns a default value to a column when no value is provided during an `INSERT` operation.
* Example: `age INT DEFAULT 18`

**Auto Increment**
* Automatically generates a unique sequential value for a column (usually for primary keys).
* Commonly used with `INTEGER` primary keys.

```sql
CREATE TABLE users (
    id INT AUTO_INCREMENT PRIMARY KEY,
);
```

**Points to remember**:
* No two rows can have the same primary keyit must be unique.
* A primary key cannot contain null values (it can't be empty).
* Each table can have only one primary key.

# Constraints Example

Below is the query to create a table student with a set of constraints.

```sql
     CREATE TABLE  student(
     student_id INTEGER PRIMARY KEY,
     student_Name TEXT UNIQUE,
     department TEXT NOT NULL CHECK (Department IN ('CSE', 'ECE', 'EEE')); 
```


* **PRIMARY KEY** can be utilized to uniquely identify a row in a table. When attempting to insert a row with the same value as an existing row in the table, a constraint violation will occur, preventing the insertion of the new row.

**UNIQUE** columns contain distinct values for each row, similar to **"PRIMARY KEY"** columns, but unlike primary key columns, a table can have multiple unique columns.

**NOT NULL** columns must have a value assigned to them. When attempting to insert a row without providing a value for a **"NOT NULL"** column, a constraint violation will occur, preventing the insertion of the new row.

A **CHECK** constraint is used to enforce specific conditions on column values in a table. It ensures that only valid data that meets the defined condition can be inserted or updated in the column. If a value violates the condition, the database rejects the operation.

**Example**: `department TEXT CHECK (department IN ('CSE', 'ECE', 'EEE'))`
* This ensures that the department column can only contain the values 'CSE', 'ECE', or 'EEE'. 
* Any attempt to insert or update a value outside this list will result in an error.
